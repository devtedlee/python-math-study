```python
%matplotlib inline
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5, 6, 7]
y = [64.3, 63.8, 63.6, 64.0, 63.5, 63.2, 63.1]

plt.plot(x, y, marker='o', color='red')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = list(range(1, 11))
y = []
for i in range(len(x)):
    y.append(3 * x[i] - 24)

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1.0, 1.01, 0.01)
y = 3 * x * x * x

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
from sympy import Symbol, solve

a = Symbol('a')
b = Symbol('b')
ex1 = a + b - 1
ex2 = 5 * a + b - 3

print(solve((ex1, ex2)))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-1, 6)
y = 1 / 2 * x + 1 / 2
y2 = -2 * x + 7

plt.plot(x, y)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
from sympy import Symbol, solve

x = Symbol('x')
y = Symbol('y')

ex1 = -3 / 2 * x + 6 - y
ex2 = 1 / 2 * x + 2 - y
print(solve((ex1, ex2)))
```
