```python
from sympy import Symbol, solve

a = Symbol('a')
b = Symbol('b')
ex1 = a + b - 1
ex2 = 5 * a + b - 3

print(solve((ex1, ex2)))
```

```python
from sympy import Symbol, solve

x = Symbol('x')
y = Symbol('y')

ex1 = -3 / 2 * x + 6 - y
ex2 = 1 / 2 * x + 2 - y
print(solve((ex1, ex2)))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a1 = (5 - 1) / (6 - 0)
b1 = 1

cx = (0 + 6) / 2
cy = (1 + 5) / 2

a2 = -1 / a1

b2 = cy - a2 * cx

x = np.arange(0, 7)
y1 = a1 * x + b1
y2 = a2 * x + b2

plt.plot(x, y1)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

th = np.arange(0, 360)

r = 5
x = r * np.cos(np.radians(th)) + 3
y = r * np.sin(np.radians(th)) - 1

plt.plot(x, y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

rad = np.arctan2(3, 4) # 높이 = 3, 밑변 = 4 인 삼각형의 각도 구하기
print(rad) # 호도법
th = np.degrees(rad)
print(th) # 각도법
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

r = 3000
x = np.arange(-r, r + 1)
y = np.sqrt(r**2 - x**2)

plt.plot(x, y)
plt.plot(x, -y)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a = 200
b = 300

r = 300
x = np.arange(a - r, a + r + 1)
y = np.sqrt(r**2 - (x - a)**2) + b
y2 = -y + 2 * b

plt.plot(x, y)
plt.plot(x, y2)
plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```
